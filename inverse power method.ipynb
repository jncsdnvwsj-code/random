```python
import numpy as np

A = np.array([[0, 2], 
              [2, 3]])
x_0 = np.array([1, 1])
s = 2
N = 100
A_s = A - s * np.identity(2)

def inverse_power_method():
    x = x_0
    for i in range(N):
        u = x / np.sqrt(x[0]**2 + x[1]**2)
        x = np.linalg.solve(A_s, u)
        R = np.matmul(np.transpose(u), x)
    return 1/(R + s) , x / np.sqrt(x[0]**2 + x[1]**2)

print(inverse_power_method())
```
